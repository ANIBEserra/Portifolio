```python
import cloudscraper
import pandas as pd

scraper = cloudscraper.create_scraper() # Cria um scraper que simula um browser real
url = "https://fbref.com/en/comps/9/shooting/stats/Premier-League-Stats"

response = scraper.get(url)

if response.status_code == 200:
    df = pd.read_html(response.text)[0]
    df.columns = df.columns.droplevel(0)
    print("FINALMENTE! Conseguimos os dados.")
    print(df[['Player', 'Sh', 'SoT']].head())
else:
    print(f"Bloqueio persistente: {response.status_code}")
```

```python
import cloudscraper
import pandas as pd

scraper = cloudscraper.create_scraper()

# URL principal do time (Geralmente mais estável que os logs detalhados)
url = "https://fbref.com/en/squads/05439c82/Bayern-Munich-Stats"

print("Tentando acesso à página principal do Bayern...")
response = scraper.get(url)

if response.status_code == 200:
    # A primeira tabela (index 1 ou 2) geralmente é a de 'Scores & Fixtures'
    # que já contém Chutes e Faltas em colunas resumidas
    dfs = pd.read_html(response.text)
    
    # Procurando a tabela de resultados (Scores & Fixtures)
    df_results = dfs[1] 
    
    print("Sucesso! Pegamos a tabela de resultados.")
    # Exibindo as colunas disponíveis para validarmos o que veio
    print(df_results[['Date', 'Opponent', 'Result']].tail(5))
else:
    print(f"Erro {response.status_code}. O servidor ainda está recusando.")
```
